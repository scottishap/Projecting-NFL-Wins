```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dvoa_2010 = pd.read_csv('./dvoa_2010.csv',index_col='Team',usecols=['Team','Total DVOA','Offense DVOA','Defense DVOA','ST DVOA','WtdDVOA','Est. Wins']).add_suffix('_2010')
dvoa_2011 = pd.read_csv('./dvoa_2011.csv',index_col='Team',usecols=['Team','Total DVOA','Offense DVOA','Defense DVOA','ST DVOA','WtdDVOA','Est. Wins']).add_suffix('_2011')
dvoa_2012 = pd.read_csv('./dvoa_2012.csv',index_col='Team',usecols=['Team','Total DVOA','Offense DVOA','Defense DVOA','ST DVOA','WtdDVOA','Est. Wins']).add_suffix('_2012')
dvoa_2013 = pd.read_csv('./dvoa_2013.csv',index_col='Team',usecols=['Team','Total DVOA','Offense DVOA','Defense DVOA','ST DVOA','WtdDVOA','Est. Wins']).add_suffix('_2013')
dvoa_2014 = pd.read_csv('./dvoa_2014.csv',index_col='Team',usecols=['Team','Total DVOA','Offense DVOA','Defense DVOA','ST DVOA','WtdDVOA','Est. Wins']).add_suffix('_2014')
dvoa_2015 = pd.read_csv('./dvoa_2015.csv',index_col='Team',usecols=['Team','Total DVOA','Offense DVOA','Defense DVOA','ST DVOA','WtdDVOA','Est. Wins']).add_suffix('_2015')
dvoa_2016 = pd.read_csv('./dvoa_2016.csv',index_col='Team',usecols=['Team','Total DVOA','Offense DVOA','Defense DVOA','ST DVOA','WtdDVOA','Est. Wins']).add_suffix('_2016')

```

```python
team_dict = {'ARI':'Arizona Cardinals',
'ATL':'Atlanta Falcons',
'BAL':'Baltimore Ravens',
'BUF':'Buffalo Bills',
'CAR':'Carolina Panthers',
'CHI':'Chicago Bears',
'CIN':'Cincinnati Bengals',
'CLE':'Cleveland Browns',
'DAL':'Dallas Cowboys',
'DEN':'Denver Broncos',
'DET':'Detroit Lions',
'GB':'Green Bay Packers',
'HOU':'Houston Texans',
'IND':'Indianapolis Colts',
'JAC':'Jacksonville Jaguars',
'KC':'Kansas City Chiefs',
'MIA':'Miami Dolphins',
'MIN':'Minnesota Vikings',
'NE':'New England Patriots',
'NO':'New Orleans Saints',
'NYG':'New York Giants',
'NYJ':'New York Jets',
'OAK':'Oakland Raiders',
'PHI':'Philadelphia Eagles',
'PIT':'Pittsburgh Steelers',
'SD':'San Diego Chargers',
'SEA':'Seattle Seahawks',
'SF':'San Francisco 49ers',
'STL':'St. Louis Rams',
'TB':'Tampa Bay Buccaneers',
'TEN':'Tennessee Titans',
'WAS':'Washington Redskins'}
```

```python
dvoa_2010.index = dvoa_2010.index.map(lambda x: team_dict[x])
dvoa_2011.index = dvoa_2011.index.map(lambda x: team_dict[x])
dvoa_2012.index = dvoa_2012.index.map(lambda x: team_dict[x])
dvoa_2013.index = dvoa_2013.index.map(lambda x: team_dict[x])
dvoa_2014.index = dvoa_2014.index.map(lambda x: team_dict[x])
dvoa_2015.index = dvoa_2015.index.map(lambda x: team_dict[x])
dvoa_2016.index = dvoa_2016.index.map(lambda x: team_dict[x])
```

```python
df_new = pd.concat([dvoa_2010,dvoa_2011,dvoa_2012,dvoa_2013,dvoa_2014,dvoa_2015,dvoa_2016],axis=1,join='inner')
```

```python
#build three tear looks 2010-2012

df_2010_2012 = df_new[[i for i in df_new.columns if ('2010' in i) | ('2011' in i)|('2012' in i)]]

df_2010_2012.index = [i + ' (2010-2012)' for i in df_2010_2012.index]

df_2010_2012.columns = df_2010_2012.columns.map(lambda x: x.replace('2010','py_2').replace('2011','py_1').replace('2012','py').replace('2013','next'))

#build three tear looks 2011-2013
df_2011_2013 = df_new[[i for i in df_new.columns if ('2011' in i) | ('2012' in i)|('2013' in i)]]

df_2011_2013.index = [i + ' (2011-2013)' for i in df_2011_2013.index]

df_2011_2013.columns = df_2011_2013.columns.map(lambda x: x.replace('2011','py_2').replace('2012','py_1').replace('2013','py').replace('2014','next'))

#build three year looks 2012-2014
df_2012_2014 = df_new[[i for i in df_new.columns if ('2012' in i) | ('2013' in i)|('2014' in i)]]

df_2012_2014.index = [i + ' (2012-2014)' for i in df_2012_2014.index]

df_2012_2014.columns = df_2012_2014.columns.map(lambda x: x.replace('2012','py_2').replace('2013','py_1').replace('2014','py').replace('2015','next'))

#build three year looks 2013-2015

df_2013_2015_adv = df_new[[i for i in df_new.columns if ('2013' in i) | ('2014' in i)|('2015' in i)]]

df_2013_2015_adv.index = [i + ' (2013-2015)' for i in df_2013_2015_adv.index]

df_2013_2015_adv.columns = df_2013_2015_adv.columns.map(lambda x: x.replace('2013','py_2').replace('2014','py_1').replace('2015','py').replace('2016','next'))

df_three_year_adv = pd.concat([df_2010_2012,df_2011_2013,df_2012_2014])

print df_three_year_adv.shape
df_three_year_adv.head()
```
