To look at a different benchmark, how much does my model vary from a more establish projection. Notable analytical football site, football outsiders, released their mean projections going into season based on simulations they ran. Below is the difference and correlation between my projections and theirs. 

```python
import pandas as pd
```

```python
wins_2017 = pd.read_csv('2017_proj_wins.csv')
```

```python
wins_2017.drop('Rk',axis=1,inplace=True)
```

```python
wins_2017.sort_values('2017 wins',ascending=False)
```

```python
wins_2017.corr()
```

```python
import scipy.stats as stats
```

```python
stats.spearmanr(wins_2017['2017 wins'],wins_2017['FO_wins'])[0]
```

```python
wins_2017['win_diff'] = wins_2017['2017 wins'] - wins_2017['FO_wins']
```

```python
wins_2017.sort_values('FO_wins')
```

```python
Glancing at the teams that stand out, this confirms my assumption that I have failed to quantify the offseason. 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
plt.figure(figsize=(8,8))
plt.scatter(wins_2017['FO_wins'],wins_2017['2017 wins'])
plt.plot(wins_2017['FO_wins'],wins_2017['FO_wins'])
plt.title('2017 Win Predictions vs 2017 Football Outsiders',fontsize=18)
plt.xlabel('Model Wins',fontsize=16)
plt.ylabel('Football Outsider Wins',fontsize=16)
```

```python
wins_2017[['team','2017 wins','FO_wins']].sort_values('2017 wins')
```

```python
wins_2017['model_rank'] = wins_2017[['2017 wins']].rank(ascending=False)
```

```python
wins_2017['FO_rank'] = wins_2017[['FO_wins']].rank(ascending=False)
```

```python
wins_2017['rank_diff'] = wins_2017['FO_rank'] - wins_2017['model_rank']
```

```python
wins_2017.sort_values('rank_diff')
```
